## Fala para texto

```python
import speech_recognition as sr
import os
import playsound
import webbrowser
import wikipedia
import pyaudio
from datetime import datetime
```

```python
def speak(text):
    tts = gTTS(text=text, lang='pt', tld='com.br')
    filename = "voice.mp3"
    try:
        os.remove(filename)
    except OSError:
        pass
    tts.save(filename)
    playsound.playsound(filename)
```

```python
def get_audio():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        r.pause_threshold = 1
        r.adjust_for_ambient_noise(source, duration=1)
        audio = r.listen(source)
        said = ""
        try:
            said = r.recognize_google(audio)
            print(said)
        except sr.UnknownValueError:
            speak("Fale novamente, não entendi.")
        except sr.RequestError:
            speak("O serviço não está disponível")
    return said.lower()
```

```python
def respond(text):
    print("fala para texto " + text)
    if 'youtube' in text:
        speak("O que você quer pesquisar?")
        keyword = get_audio()
        if keyword!= '':
            url = f"https://www.youtube.com/results?search_query={keyword}"
            webbrowser.get().open(url)
            speak(f"Aqui está o que encontrei para {keyword} no YouTube")
    elif 'pesquisa' in text:
        speak("O que você quer pesquisar?")
        query = get_audio()
        if query !='':
            result = wikipedia.summary(query, sentences=3)
            speak("Resultado da wikipedia")
            print(result)
            speak(result)
    elif 'horas' in text:
        strTime = datetime.today().strftime("%H:%M %p")
        print(strTime)
        speak(strTime)
    elif 'sair' in text:
        speak("Adeus, até a próxima")
        exit()
```

```python
while True:
    print("I am listening...")
    text = get_audio()
    respond(text)
```
